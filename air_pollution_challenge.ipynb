```python
X , Y = data[:,:-1],data[:,5]
print(X.shape,Y.shape)
```

```python
m=X.shape[0]
X0 = np.ones([m,1])
X = np.hstack([X0,X])
print(X)
```

```python
print(X.shape)
```

```python
print(type(Y))
```

```python
Y =Y.reshape((-1,1))
```

```python
theta = np.dot(np.linalg.pinv(np.dot(X.T,X)),np.dot(X.T,Y))
```

```python
print(theta.shape)
```

```python
theta = theta.reshape((-1,))
print(theta.shape)
print(theta)
```

```python
def hypothesis(x,theta):
    y=0
    m = 6
    for i in range(m):
        y += x[i]*theta[i]
    return y
```

```python
m= x_test.shape[0]
X0 = np.ones([m,1])
x_test = np.hstack([X0,x_test])
print(x_test)
```

```python
y_test = []
for i in range(x_test.shape[0]):
    y_test.append(hypothesis(x_test[i,:],theta))

y_test = np.array(y_test)
y_test.reshape(-1,1)
print(y_test.shape)
```

```python
with open('Sample_Submission.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    m = y_test.shape[0]
    writer.writerow(["Id","Target"])
    for i in range(m):
        l = [i,y_test[i]]
        writer.writerow(l)
    csvFile.close()
```
